```python
# Sistema simples de clínica

pacientes = []
consultas = []

def cadastrar_paciente():
    nome = input("Digite o nome do paciente: ")
    idade = input("Digite a idade: ")
    
    paciente = {
        "nome": nome,
        "idade": idade
    }
    
    pacientes.append(paciente)
    print("Paciente cadastrado com sucesso!\n")

def listar_pacientes():
    if not pacientes:
        print("Nenhum paciente cadastrado.\n")
        return
    
    print("Lista de pacientes:")
    for i, p in enumerate(pacientes):
        print(f"{i + 1}. Nome: {p['nome']} | Idade: {p['idade']}")
    print()

def agendar_consulta():
    if not pacientes:
        print("Cadastre um paciente primeiro.\n")
        return
    
    listar_pacientes()
    escolha = int(input("Escolha o número do paciente: ")) - 1
    
    if escolha < 0 or escolha >= len(pacientes):
        print("Opção inválida.\n")
        return
    
    data = input("Digite a data da consulta: ")
    
    consulta = {
        "paciente": pacientes[escolha]["nome"],
        "data": data
    }
    
    consultas.append(consulta)
    print("Consulta agendada com sucesso!\n")

def listar_consultas():
    if not consultas:
        print("Nenhuma consulta agendada.\n")
        return
    
    print("Consultas agendadas:")
    for c in consultas:
        print(f"Paciente: {c['paciente']} | Data: {c['data']}")
    print()

def menu():
    while True:
        print("=== Sistema de Clínica ===")
        print("1 - Cadastrar paciente")
        print("2 - Listar pacientes")
        print("3 - Agendar consulta")
        print("4 - Listar consultas")
        print("5 - Registrar transação")
        print("6 - Listar transações")
        print("7 - Ver saldo")
        print("8 - Relatório geral")
        print("9 - Relatório de desempenho")
        print("0 - Sair")
        
        opcao = input("Escolha uma opção: ")
        
        if opcao == "1":
            cadastrar_paciente()
        elif opcao == "2":
            listar_pacientes()
        elif opcao == "3":
            agendar_consulta()
        elif opcao == "4":
            listar_consultas()
        elif opcao == "5":
            registrar_transacao()
        elif opcao == "6":
            listar_transacoes()
        elif opcao == "7":
            saldo_total()
        elif opcao == "8":
            relatorio_geral()
        elif opcao == "9":
        relatorio_desempenho()    
        elif opcao == "0":
            print("Saindo...")
            break
        else:
            print("Opção inválida!\n")

menu()
```
